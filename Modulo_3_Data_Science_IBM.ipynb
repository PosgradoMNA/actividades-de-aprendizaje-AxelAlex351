# ANOVA

El análisis de varianza es un análisis estadístico que hace uso de la varianza para estudiar ciertos componentes debidos a diferentes variables explicativas. El ANOVA puede ser utilizado para encontrar la correlación entre diferentes grupos de una variable categórica. La prueba de ANOVA regresa dos valores: F-test score y p-value.

La prueba-F calcula la proporción de la variación entre las medias de grupos sobre la variación dentro de cada grupo de la muestra. El valor-p muestra si el resultado obtenido es significativo estadísticamente (nivel de confianza). Para utilizar el ANOVA en python, se hace uso del método **f_oneway** de la paquetería Scipy. 

#Correlación

La correlacióm es una métrica estadística para medir la extensión de la interdependencia de las variables, en otras palabras si tenemos dos variables a lo largo del tiempo, si una de las variables cambia, entonces ¿cómo dicho cambio afecta a la otra variable? Es la conexión entre estas variables. Por ejemplo, fumar está fuertemente correlacionado con el cáncer de pulmón, entonces un individuo que fuma tiene mayor probabilidad de padecer de está enfermedad. Es importante remarcar que la correlación no significa causalidad, un ejemplo de esto es la lluvia y las sombrillas a más precipatación, mayor serán el número de sombrillas que la gente portará de igual forma sino llueve casi nadie usará sombrillas; sin embargo; no se tiene la información de si portar una sombrilla es causante de la precipitación o si el simple hecho de llover causa que aparezca una sombrilla. Para checar la correlación en las variables dentro de un DataFrame se utiliza el método gráfico con la paquetería seaborn **regplot** gráfico de regresión, cuyo objetivo es ver si una variable tiene impacto en la otra. 

#Correlación Pearson

Una forma de medir la intensidad de la correlación entre variables numéricas contínuas es usando el método de correlación de Pearson. La correlación de Pearson, da dos valores, el coeficiente de correlación y el valor-p. 

Estos valores se interpretan de la siguiente forma: para el coeficiente de correlación, si el valor es cercano a 1, esto implica que se tiene una correlación alta positiva, mientras que si el valor es cercano a -1 entonces esto indica que existe una alta correlación negativa, cuando el valor es cercano a 0, significa que no existe correlación alguna entra las variables. 
El valor-p muestra la certeza de dicha correlación. Para un valor-p menor a 0.001 nos da una fuerte certidumbre acerca del coeficiente de correlación calculado. Para el rango de 0.001 < p-value < 0.05 la certidumbre del coeficiente de correlación calculado es moderada, entre 0.05 y 0.1 se genera una certidumbre débil y un valor mayor de 0.1 no nos genera ningún tipo de certidumbre. Por lo tanto, se dice que existe una correlación alta entre las variables de estudio si estás presentan un coeficiente de correlación cercano a 1 o -1, y si el valor-p es < 0.001. Para calcular la correlación se utiliza el método **.pearsonr()** de la paquetería Scipy.
