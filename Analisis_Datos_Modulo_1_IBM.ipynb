**Adquisición de datos** es el proceso de cargar y leer los datos en un cuaderno de varias fuentes. Para leer cualquier tipo de dato usando Python, la mejor paquetería es *pandas*, hay que considerar dos factores importantes:

1. El formato del archivo
2. La dirección del archivo (file path)

El formato es el modo en que los datos están codificados, para hallar el formato tenemos que ver el final del archivo, los más comunes terminan con: **.csv, .json, .xlsx, .hdf,** etc. La dirección nos dice en donde se encuentra almacenado dicho archivo, usualmente está almacenado ya sea en la computadora o en internet. Un ejemplo de dirección: Computadora: /Desktop/mydata.csv, internet: https://...

Para el caso de Tomás, para extraer la información del sitio web de interés:
https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data
este archivo en realidad no significa nada porque está codificado en formato .csv, para poder extraer la información utilizaremos la librería de pandas para poder interpretarlo:

```python
# En pandas el método read_csv() puede leer los archivos con las columnas separadas por comas a un dataframe. 
import numpy as np
import pandas as pd # importamos pandas y lo asignamos a pd
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data' # definimos variable con la dirección del archivo
df = pd.read_csv(url) # usamos el método de pandas para decodificar y arreglar datos
df #desplegamos la variable dataframe
```

No damos cuenta que al usar el método: *read_csv* este asume que todos los datos contienen un encabezado, y en los datos cargados la columna de autos usados por marca no hay un encabezadon entonces podemos definir dentro de la función de encabezados del método, la asignación de *None*.

```python
df = pd.read_csv(url, header = None)
df.head() # Enseña las primeras 5 filas o (n) filas del data frame
# df.tail(n) enseña las últimas n filas del data frame (por default son 5)
```

```python
headers = ['symboling','normalized-losses','make', 'fuel-type','aspiration', 'num-of-doors','body-style','drive-wheels','engine-location','wheel-base','length','width', 'height','curb-weight','engine-type','number-of-cylinders','engine-size','fuel-system','bore','stroke','compression-ratio','horsepower','peak-rpm','city-mpg','highway-mpg','price' ]
df.columns = headers # el método .columns selecciona del df todas las columnas y las renombra con la variable de encabezados
df.head()
```

Reemplazamos los valores de '?'con el símbolo NaN o not a number, utilizamos el método: '*dropna()*'

```python
df1 = df.replace('?',np.NaN) # Reemplazamos los ? por el numpy NaN
df1.head()
```

Matamos los valores faltantes con dropna() en la columna de precios:

```python
df = df1.dropna(subset=['price'], axis=0) # 0 para filas, 1 para columnas 
df.head(20)
```

```python
df.dtypes # te da el tipo de datos de cada columna 
```

Si quisieramos sacar un resumen estadístico de variables tipo numéricas (int, y float) de cada columna, entonces el método: .describe() nos genera el resumen.
