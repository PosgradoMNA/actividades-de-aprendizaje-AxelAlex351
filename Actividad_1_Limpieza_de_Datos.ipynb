Solamente la columna con la variable $x_1$ no presenta datos perdidos o tipo NaN y None. $x_1$ representa la cantidad dada de crédito (Amount of the given credit)

```python
dfp = df.copy() # Se crea copia y se guarda en variable 'data frame prime'
dfp.sort_values(by = ['X1','X2','X3','X4','X5','X6','X7','X8','X9','X10','X11','X12','X13','X14','X15','X16','X17','X18','X19','X20','X21','X22','X23'], ascending=True)
```

```python
dfp.dropna(axis = 0,thresh = 12 , inplace = True) # Claramente si utilizamos axis = 1 sólo la primer columna sobrevivirá
dfp # Si hay más de 12 Nan's la fila se descartará para evitar el ruido
```

```python
dfp.isna().any()
```

Para definir como rellenar los datos faltantes es necesario conocer los valores extremos o máximos y mínimos de cada una de las columnas, para esto hay que barrer columnas y preguntar el valor máximo y mínimo de dicha lista. 

```python
contador = 0
for column in dfp:
  X = dfp[column]
  contador = contador + 1
  print(contador,'Valor máximo:', max(X), 'Valor mínimo:', min(X))

```

Nos damos cuenta que $x_1$, $x_2$, $x_3$, $x_4$, y $x_6$. Ya fueron limpiadas con la condición de eliminar las filas que tuvieran al menos 10 NaN's

$x_5$: Age, es la edad de la persona que, nos damos cuenta que la mayor tiene 79 años y la menor 21, pero yo consideraré el promedio como datos faltantes, puesto que sí hacemos la diferencia $\Delta = max-min$ = 58 años y el medio de la diferencia $\frac{\Delta}{2}$ = 29 años. Como percibo que no hay tanto margen entre estos dos extremos pienso que usar el promedio que es 35.5 años no producirá tanto sesgo. 

```python
agemean = dfp.X5.mean()
agemean
```

```python
dfp['X5'].fillna(value = agemean, inplace = True)
dfp['X5'].isna().any()
```

```python
dfp.isna().any()
```

De las variables $x_6$ - $x_{11}$ son el historial de pagos **History of past payment** la clave indica que por cada mes que no se pagaba se añadía una unidad el mes empieza con Septiembre de 2005 en $x_6$, Octubre $x_7$ y así sucesivamente, a los que pagan debidamente se les denota con un $-1$, los que no se acumula una unidad por cada mes faltante.

Para no complicar tanto el ejercicio sólo reemplazaré cada uno de los datos faltantes por su mediana en cada columna. 

```python
sec1 = dfp.loc[:,'X6':'X11']
```

```python
sec1.describe()
```

```python
for i in sec1: # Checamos las medianas
  mX = dfp[i].median()
  dfp[i].fillna(value = mX, inplace = True)
  print(mX)
```

```python
dfp.isna().any()
```

De igual forma se checan las medianas de los otros valores que son la cantidad de pagos de adeudo y los pagos previos, notamos que los mínimos varían mucho de los máximos, no tiene caso sacar el promedio.

```python
sec2 = dfp.loc[:,'X12':'X23']
sec2
```

```python
for i in sec2: # Checamos las medianas
  S2 = dfp[i].median()
  dfp[i].fillna(value = S2, inplace = True)
  print(S2)
```

```python
dfp.isna().any()
```
