#Listas y Tuplas

Estos son llamados tipos de datos compuestos.

**Tuplas**- son una secuencia ordenada, se expresan mediante elementos separados por comas dentro de un paréntesis. 

En python las tuplas pueden contener strings, enteros, y flotantes, y todas pueden estar dentro de una sola tupla, pero el tipo de variable será *tupla*.


```python
Tuple_1 = ('CD',100,23.45)
type(Tuple_1)
```

Para acceder a los índices de los carácteres que conforman la tupla se utiliza el formato del nombre de la variable que se le asigna la tupla seguido de unos brackets [  ].

```python
Tuple_1 = ('CD',100,23.45)
Tuple_1[0] # Nos debe de retornar completamente el primer elemento
```

Se pueden usar índices negativos:

image.png

De igual forma se pueden concatenar las tuplas mediante el operador de adición +.

```python
Tuple_2 = Tuple_1 + ('DVD', 89) 
Tuple_2
```

Se pueden cortar tuplas. El último índice es una posición más adelante que el índice que se quiere:

('disco', 10, 20.4, 'CD', 9)

tupla_3[0:3] → disco, 10, 20.4 (slicing)

El largo de la tupla se puede reconocer mediante len().

Las tuplas son inmutables, es decir, no se pueden cambiar, al asignar una tupla dentro de una variable *x*, y *x* asignarla dentro de otra llamada *x_prime* no se pueden realizar operaciones en la nueva variable *x_prime* porque ambas tienen como referencia a la misma tupla. Entonces si uno desea cambiar un elemento de la tupla *x_prime*, dicha operación no se realizará, y tampoco *x* será cambiada.

Si uno desea ordenar o clasificar una tupla, uno debe de usar el comando **sorted()** y guardar esa tupla en otra variable, ejemplo:

```python
Tupla_3 = (2,3,7,8,10,10,15,12,9,4,1)
Tupla_4 = sorted(Tupla_3)
Tupla_4
```

Las tuplas pueden contener tuplas dentro de ellas mismas, a esto se le conoce en inglés como **nesting** sucede lo mismo con los ciclos. Al anidar una tupla dentro de otra tupla, para extraer dichas tuplas se usa el mismo comando de corte. Ej:

NT = (1,2,('baile','danza'),(7.8,9,111),(10,9))

Claramente podemos observar que el largo de la tupla es de 5 caracteres que la conforman, por lo que para extraer danza, usamos **NT[2]** → ('baile','danza')[1] → danza

```python
NT = (1,2,('baile','danza'),(7.8,9,111),(10,9))
NT[2][1] #árbol
```

Se puede acceder a más elementos dentro de las tuplas mediante los árboles, cada bracket te puede brindar mayor elementos por cada operador que utilices dentro de la variable, si quisieramos sacar la letra a de danza podemos hacer:

```python
NT[2][1][4]
```

Las **listas** son también una secuencia ordenada de elementos, la diferencia radica en que las listas utilizan brackets en vez de paréntesis y son mutables. Pueden contener, strings, flotantes, enteros, listas e incluso tuplas. De igual manera se acceden mediante índices. 

```python
L = ['As', 10.4, 56]
L[1]
```

Se pueden cortar las listas mediante slicing, de igual forma que las tuplas. Para esto es recordar que se incluye el índice límite inferior y el límite superior es + 1.

```python
L[1:3]
```

Se pueden concatenar las listas mediante los mismos operadores de adición:

```python
L + ['tres',2.0,1]
```

Al ser mutables se pueden añadir dentro de la misma lista nuevos elementos, sin necesidad de renombrar una nueva variable, al usar el comando **.extend( )**:

```python
L = ['As', 10.4, 56]
L.extend(['Lima',78.3])
L
```

Otro comando utilizado es **.append( )** que añade totalmente una lista nueva dentro de la lista, en vez de añadir elemento por elemento como extend

```python
L.append(['cinco',5])
L
```

Para cambiar un elemento de la lista se hace referencia a su índice y se modifica:

```python
L[0] = 'Asteroid'
L
```

De igual manera mediante el comando **del()** podemos eliminar un elemento de la lista:

```python
del(L[5])
L
```

Se puede convertir un string a formato de lista mediante el comando **.split( )**

```python
'rock and roll'.split()
```

Igualmente se pueden separar los strings individuales por medio de un carácter denominado delimitador dentro del argumento del comando split(arg)

```python
'C,G,Amin,F'.split(',') # Progresión de acordes en do mayor
```

Las listas al ser referenciadas dentro de otra variable sí comparten la primera como la última las modificaciones, a esto se le conoce como alias.

Para evitar esto, se puede clonar la variable inicial mediante **[:]**

```python
A = ['C', 'G', 'Amin', 'F']
B = A[:]
B[3] = 'Dmin'
B,A
```

#Conjuntos

Son un tipo de colección, esto significa que se parecen a las listas u las tuplas pero en desorden (no graban la posición de elemento). Solamente poseen elementos únicos, es decir, sólo existe un solo elemento dentro de este conjunto no repetible. Se usan las llaves para definir los conjuntos:

```python
Set1 = {'rock','pop','blues','R&B','soul','blues','rock','blues'}
Set1
```

Se puede pasar de un conjunto a una lista mediante la función **set( )**

```python
Lista1 = ['Hola','adiós','Hola','Hello','Salut']
Conj1 = set(Lista1)
Conj1
```

Se pueden realizar operaciones en conjuntos por medio de funciones como **.add( )** - la cual añade un elemento a la lista. **.remove( )** para eliminar el elemento que se desea remover. El comando **in** te revela si el elemento existe dentro de un conjunto, regresará respuesta booleana, T o F. Los sets se representan por medio de los diagramas de *Venn*. Y como en matemáticas cuando queremos determinar las intersecciones de un conjunto se utiliza el comando de ampersand **&** para denotar dicha intersección: 

$$A \cap B $$

De igual forma podemos encontrar la unión de todos los elementos del conjunto mediante **.union( )**

$$A \cup B $$

E igual se puede checar uno de estos diagramas es subconjunto de otro conjunto mediante la función **.issubset( )**

```python
s1 = {'Batman','Superman','Wonderwoman','Hulk','Thor'} #superheroes DC y Marvel
s2 = {'Ironman','Spiderman','Superman','Hulk','Flash'}
s1.add('Robbin')
s1
```

```python
s1.remove('Thor')
s1
```

```python
'Ironman' in s2
```

```python
s1 & s2
```

```python
s1.union(s2)
```

#Diccionarios

Los diccionarios son un tipo de colección en Python, a diferencia de las listas o las tuplas, los diccionarios poseen ***llaves*** (key) ánalogamente a los *índices* (index) y en vez de elementos, se usan valores. Las llaves pueden ser no enteros, usualmente son caractéres.

Para crear diccionarios se utilizan las llaves { } como los conjuntos. Las llaves son inmutables y únicas. 

```python
# ejemplo de diccionario
{'key1':1,'key2':'2','key3':[3,3,3],'key4':(4,4,4),('key5'):5}
```

Cada llave está acompañada de un valor separado por medio de dos puntos **:** y cada par de llave y valor separados a su vez entre comas. Estos valores pueden ser inmutables, mutables o duplicados.

Como ejemplo podemos tomar álbumes de música y el año en el que emergieron:

```python
Dict = {'Thriller':1982, 'The Dark Side of the Moon':1973,'Back in Black':1980,'The Bodyguard':1992}
```

A esta variable se le puede buscar un valor usando los brackets [ ]; El argumento es la llave, esto brindará la salida del valor. 

```python
Dict['Thriller']
```

```python
Se pueden añadir nuevas entradas al diccionario y eliminar entradas:
```

```python
Dict['Rumors'] = 1977
Dict
```

```python
del(Dict['Thriller'])
Dict
```

Verificar si hay un elemento dentro del diccionario mediante la función in

```python
'The Bodyguard' in Dict
```

Se puede utilizar el método de **.keys( )** para obtener las llaves. La salida es una lista como objeto con todas las llaves, y de igual forma podemos obtener todos los valores mediante el método **.values( )**

```python
Dict.keys()
```

```python
Dict.values()
```
